```python
import pandas as pd
data = pd.read_csv("preprocess_4.csv")
data['date'] = pd.to_datetime(data['date'])
data.head()
```

```python
unneeded = [
"is_night_game"
,"home_team_rest"
,"away_team_rest"
,"home_pitcher_rest"
,"away_pitcher_rest"
,"home_pitcher"
,"away_pitcher"
,"home_team_season"
,"away_team_season"]

for feature in unneeded:
    data = data.drop(feature, axis = 1,  inplace = True)
```
