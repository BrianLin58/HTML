```python
# delete unneeded features
# is_night_game
# home_team_rest away_team_rest home_pitcher_rest away_pitcher_rest
# home_pitcher away_pitcher
# home_team_season away_team_season
```

```python
import pandas as pd
data = pd.read_csv("preprocess_4.csv")
data['date'] = pd.to_datetime(data['date'])
data.head()
```

```python
unneeded_features = [
"id"    
,"home_team_abbr"
,"away_team_abbr"
,"date"
,"is_night_game"
,"season"
,"home_team_rest"
,"away_team_rest"
,"home_pitcher_rest"
,"away_pitcher_rest"
,"home_pitcher"
,"away_pitcher"
,"home_team_season"
,"away_team_season"]

data.drop(unneeded_features, axis = 1,  inplace = True)
```

```python
prefix = ["home_", "away_"]
suffix = ["_std", "_skew"]
Team_Seasonal = [
"team_errors"
,"team_spread"
,"team_wins"
,"batting_batting_avg"
,"batting_onbase_perc"
,"batting_onbase_plus_slugging"
,"batting_leverage_index_avg"
,"batting_wpa_bat"
,"batting_RBI"
,"pitching_earned_run_avg"
,"pitching_SO_batters_faced"
,"pitching_H_batters_faced"
,"pitching_BB_batters_faced"
,"pitching_leverage_index_avg"
,"pitching_wpa_def"]

for pre in prefix:
    for suf in suffix:
        for TS in Team_Seasonal:
            TS = pre + TS + suf
            data.drop(TS, axis = 1,  inplace = True)
```

```python
prefix = ["home_", "away_"]
suffix = ["_std", "_skew"]
Pitcher_Seasonal = [
"pitcher_earned_run_avg"
,"pitcher_SO_batters_faced"
,"pitcher_H_batters_faced"
,"pitcher_BB_batters_faced"
,"pitcher_leverage_index_avg"
,"pitcher_wpa_def"]

for pre in prefix:
    for suf in suffix:
        for PS in Pitcher_Seasonal:
            PS = pre + PS + suf
            data.drop(PS, axis = 1,  inplace = True)
```

```python
data.to_csv("preprocess_5.csv", index = False)
```

```python
with open("preprocess_5.csv", "r") as IF:
    with open("training.txt", "w") as OF:
        IF.readline()
        for l in range(data.shape[0]):
            x = IF.readline().strip().split(",")
            
            if x[0] == "True":
                x[0] = "1"
            else:
                x[0] = "-1"
            
            for i in range(1, len(x)):
                x[i] = str(i) + ":" + x[i]
            
            x = (" ").join(x)
            OF.writelines(x + "\n")
```
